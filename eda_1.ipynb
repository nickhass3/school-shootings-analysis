# EDA

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read in datasets
df = pd.read_parquet('data/clean/school_shootings.parquet')
state_shootings = pd.read_csv('data/state_shootings.csv')
```

```python
# plot the number of shootings per year
df['Date'].dt.year.value_counts().sort_index().plot(kind='bar', figsize=(15, 5))
plt.title('Number of School Shootings per Year')
plt.xlabel('Year')
plt.ylabel('Number of Shootings')
plt.show()
```

```python
# plot the number of shootings per year with a line plot
df['Date'].dt.month.value_counts().sort_index().plot(kind='bar', figsize=(15, 5))
plt.title('Number of School Shootings by Month')
plt.xlabel('Month')
plt.ylabel('Number of Shootings')
plt.show()
```

Very little shootings in the summer months. School is usually not conducted in summer months.

Many shootings when school starts (September)

Many shootings after the holiday season (January, February)

```python
# plot the age of the shooter
df['age_of_shooter'].value_counts().sort_index().plot(kind='bar', figsize=(15, 5))
plt.title('Age of Shooter')
plt.xlabel('Age')
plt.ylabel('Number of Shootings')
plt.show()
```

```python
state_shootings.sort_values(by='Shootings per 100k', ascending=False).plot(kind='bar', x='State', y='Shootings per 100k', figsize=(15, 5))
plt.title('Shootings per 100k by State')
```

```python
# plot the df by date, line graph
plt.scatter(df['Date'], df['Total'], s=5)
plt.title('Total deaths and injuries in School Shootings')
plt.xlabel('Date')
plt.ylabel('Total Deaths and Injuries')
plt.show()
```
